```python
def clean_dataframe(df, name):
    print(f"Cleaning {name}")
    return df.dropDuplicates().na.drop("all")

cleaned_order_df = clean_dataframe(order_df, "Orders")
display(cleaned_order_df)





    
```

##Convert Date Columns

```python
cleaned_order_df = cleaned_order_df.withColumn("order_purchase_timestamp", to_date(col("order_purchase_timestamp"), "yyyy-MM-dd HH:mm:ss"))
display(cleaned_order_df)
```

##Calculate order and time delays

```python
cleaned_order_df = cleaned_order_df.withColumn("actual_delivery_time", date_diff(col("order_delivered_customer_date"), col("order_purchase_timestamp")))
cleaned_order_df = cleaned_order_df.withColumn("estimated_delivery_time", date_diff(col("order_estimated_delivery_date"), col("order_purchase_timestamp")))
cleaned_order_df = cleaned_order_df.withColumn("delay", col("actual_delivery_time")  - col("estimated_delivery_time"))
display(cleaned_order_df)
```

```python
def remove_duplicate_columns(df):
    columns = df.columns

    seen_columns = set()
    columns_to_drop = []

    for colum in columns:
        if colum in seen_columns:
            columns_to_drop.append(colum)
        else:
            seen_columns.add(colum)
    return df.drop(*columns_to_drop)

final_df = remove_duplicate_columns(final_df)
display(final_df)
```
